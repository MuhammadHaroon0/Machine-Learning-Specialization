```python
def get_first_name(val):
    return val.split()[0]

df['first_name']=df['Payer Name'].apply(get_first_name)
df.head()
```

```python
def quality(total_bill,tip):
    if tip/total_bill>0.25: #25%
        return "Generous"
    elif tip/total_bill>0.12:
        return "Normal"
    else:
        return "Cheap"
```

```python
df['Quality']=df[["total_bill","tip"]].apply(lambda df:quality(df["total_bill"],df["tip"]),axis=1)
```

```python
df
```

```python
# df['Quality']=df[["total_bill","tip"]].apply(lambda df:quality(df["total_bill"],df["tip"]),axis=1)
# same as
# df=df.drop('Quality',1)
df['Quality']=np.vectorize(quality)(df["total_bill"],df["tip"]) # efficient
```
