```python
X=df.drop('class',axis=1)
```

```python
X
```

```python
X=pd.get_dummies(X)
```

```python
X
```

```python
y=df['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
help(GradientBoostingClassifier)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
    'learning_rate':[0.01,0.1,0.2,0.6],
    'max_depth':[3,4,5],
    'n_estimators':[50,100]
}
```

```python
gbc=GradientBoostingClassifier()
```

```python
grid=GridSearchCV(gbc,param_grid)
```

```python
grid.fit(X_train,y_train)
```

```python
preds=grid.predict(X_test)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid.best_estimator_.feature_importances_
```

```python
feats=pd.DataFrame(index=X.columns,data=grid.best_estimator_.feature_importances_,columns=['Importances'])
```

```python
feats=feats[feats['Importances']>0]
```

```python
feats
```

```python
feats=feats[feats['Importances']>0.005]
```

```python
feats
```

```python
plt.figure(figsize=(14,8),dpi=190)
sns.barplot(data=feats.sort_values('Importances'),x=feats.index,y='Importances')
plt.xticks(rotation=90);
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(classification_report(preds,y_test))
```
