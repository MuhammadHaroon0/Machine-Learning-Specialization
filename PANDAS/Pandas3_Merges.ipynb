```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],'value': [1, 2, 7, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],'value': [5, 6, 7, 8]})
```

```python
df1
```

```python
print(df2)
```

```python
df2
```

```python
pd.merge(df1,df2,how="inner",on="value") #order does'nt matter only common elements will be shown 
```

```python
pd.merge(df1,df2,how="left",on="value") #order  matter all left dataframe elements will be shown
#and those which are not present in right will be shown as nan
```

```python
pd.merge(df1,df2,how="right",on="value")#order  matter all right dataframe elements will be shown
#and those which are not present in left will be shown as nan
```

```python
pd.merge(df1,df2,how="outer",on="value")#order does'nt all  dataframe elements will be shown from left and right dfs both
```

```python
df1=df1.set_index('value')
df1
```

```python
df3=pd.merge(df1,df2,left_index=True,right_on="value",how="inner")
df3.index=range(len(df3))
df3
```
