```python
#3d(h,w,c) --> 2d(h*w,c)
```

```python
img_arr_2d=img_as_arr.reshape(h*w,c)
```

```python
img_arr_2d
```

```python
img_arr_2d.shape
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=7)
```

```python
labels=model.fit_predict(img_arr_2d)
```

```python
labels
```

```python
len(labels)
```

```python
rgb_codes=model.cluster_centers_
```

```python
rgb_codes
```

```python
rgb_codes=rgb_codes.round(0).astype(int)
```

```python
rgb_codes
```

```python
len(rgb_codes[labels])
```

```python
quantized_img=rgb_codes[labels]
```

```python
quantized_img
```

```python
quantized_img_as_3d=quantized_img.reshape(h,w,c)
```

```python
plt.imshow(quantized_img_as_3d)
```
