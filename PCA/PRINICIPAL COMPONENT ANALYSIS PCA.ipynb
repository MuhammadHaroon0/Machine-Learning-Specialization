```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaled_df=scaler.fit_transform(df)
```

```python
scaled_df.mean() #very less
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
p_comps=pca.fit_transform(scaled_df)
```

```python
sns.scatterplot(x=p_comps[:,0],y=p_comps[:,1])
```

```python
pca.explained_variance_ratio_
```

```python
np.sum(pca.explained_variance_ratio_) #just two features explaining 63 percent of variance
```

```python
res=PCA(n_components=30)
res.fit_transform(scaled_df)
```

```python
np.sum(res.explained_variance_ratio_)
```

```python
variance=[]
for i in range(1,30):
    pca=PCA(n_components=i)
    pca.fit_transform(scaled_df)
    variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
sns.lineplot(x=range(1,30),y=variance)
```
