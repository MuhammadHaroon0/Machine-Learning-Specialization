```python
# "D:\\CS\\SIDE SKILLS\\MACHINE LEARNING\\ML SPECIALIZATION\\MATPLOTLIB\\
a=np.linspace(0,10,11)
b=a**4
x=np.arange(0,10)
y=2*x
```

```python
fig=plt.figure(figsize=(5,3),dpi=150)
axes1=fig.add_axes([0,0,1,1])

axes1.plot(a,b)

axes1.set_xlim(0,11)
axes1.set_ylim(0,20000)
axes1.set_xlabel("a")
axes1.set_ylabel("b")
axes1.set_title("a=b**4")



axes2=fig.add_axes([0.25,0.25,0.2,0.2])
axes2.plot(x,y)


axes2.set_xlim(0,10)
axes2.set_ylim(0,20)
axes2.set_xlabel("x")
axes2.set_ylabel("y")
axes2.set_title("y=2x")
fig.savefig("D:\\CS\\SIDE SKILLS\\MACHINE LEARNING\\ML SPECIALIZATION\\MATPLOTLIB\\figObj.png",bbox_inches="tight")
plt.show()

```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(6,5),dpi=150)

axes[0][0].plot(x,y)
# axes[0][0].plot(x,y)

axes[0][1].plot(a,b)
axes[1][0].plot(y,x)
axes[1][1].plot(b,a)


for i in range(0,len(axes)):
    for j in range(0,2):
        axes[i][j].set_title(f"ax[{i}][{j}]")
        axes[i][j].set_xlabel("x-axis")
        axes[i][j].set_ylabel("y-axis")
          
#can set_xlim,set_ylim
fig.suptitle("Figure 1",fontsize=14)
fig.subplots_adjust(wspace=0.6,hspace=0.6)
fig.savefig("D:\\CS\\SIDE SKILLS\\MACHINE LEARNING\\ML SPECIALIZATION\\MATPLOTLIB\\subplots2.png",bbox_inches="tight")
plt.show()
# plt.tight_layout(pad=2)
```
