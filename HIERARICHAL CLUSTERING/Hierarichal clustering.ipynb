```python
df_dummies=pd.get_dummies(df.drop("name",axis=1))
```

```python
df_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaled_df=scaler.fit_transform(df_dummies)
```

```python
scaled_df
```

```python
scaled_df=pd.DataFrame(scaled_df,columns=df_dummies.columns)
```

```python
scaled_df
```

```python
plt.figure(figsize=(15,8))
sns.clustermap(scaled_df)
```

```python
plt.figure(figsize=(15,8))
sns.clustermap(scaled_df,col_cluster=False)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agg_model=AgglomerativeClustering(n_clusters=4)
```

```python
preds=agg_model.fit_predict(scaled_df)
```

```python
preds
```

```python
sns.scatterplot(df,x='mpg',y='weight',hue=preds,palette="magma")
```

```python
sns.scatterplot(df,x='mpg',y='horsepower',hue=preds,palette="magma")
```

```python
# max distance possible?
# by euclidean distance
#sqrt(no of features i.e columns)
```

```python
np.sqrt(len(scaled_df.columns))
```

```python
model=AgglomerativeClustering(n_clusters=None,distance_threshold=0) 
#distance_threshold=0 indicates there is 0 distance b/w points and so it will have the number of clusters equal to number of points
```

```python
model.fit_predict(scaled_df)
```

```python
from scipy.cluster.hierarchy import dendrogram

from scipy.cluster import hierarchy
```

```python
linkage_matrix=hierarchy.linkage(model.children_)
```

```python
linkage_matrix
```

```python
# [first_point second_point distance no_of_points_in_this_cluster]
```

```python
plt.figure(figsize=(12,8))
dendro=dendrogram(linkage_matrix)
```

```python
plt.figure(figsize=(12,8))
dendro=dendrogram(linkage_matrix,truncate_mode="lastp",p=10)
```

```python
plt.figure(figsize=(12,8))
dendro=dendrogram(linkage_matrix,truncate_mode="level",p=3)
```
