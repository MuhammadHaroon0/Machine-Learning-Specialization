```python
from svm_margin_plot import plot_svm_boundary 
```

```python
from sklearn.svm import SVC
```

```python
# help(SVC)
```

```python
svc_model=SVC(kernel='linear',C=1000)
```

```python
X=df.drop('Virus Present',axis=1)
```

```python
y=df['Virus Present']
```

```python
svc_model.fit(X,y)
```

```python
plot_svm_boundary(svc_model,X,y)
```

```python
svc_model=SVC(kernel='linear',C=0.1)  #C is inversely proportional with the number of support vectors inside the margins
svc_model.fit(X,y)
plot_svm_boundary(svc_model,X,y)
```

```python
svc_model=SVC(kernel='rbf',C=1)  #C is inversely proportional with the number of support vectors inside the margins
svc_model.fit(X,y)
plot_svm_boundary(svc_model,X,y)
```

```python
svc_model=SVC(kernel='rbf',C=0.1)  #C is inversely proportional with the number of support vectors inside the margins
svc_model.fit(X,y)
plot_svm_boundary(svc_model,X,y)
```

```python
svc_model=SVC(kernel='rbf',gamma='auto')  #gamma is for fitting precisely
svc_model.fit(X,y)
plot_svm_boundary(svc_model,X,y)
```

```python
svc_model=SVC(kernel='rbf',gamma=2)  #overfitting
svc_model.fit(X,y)
plot_svm_boundary(svc_model,X,y)
```

```python
svc_model=SVC(kernel='rbf',gamma=0.02)  #underfitting
svc_model.fit(X,y)
plot_svm_boundary(svc_model,X,y)
```

```python
svc_model=SVC(kernel='sigmoid')  #not optimum for this dataset
svc_model.fit(X,y)
plot_svm_boundary(svc_model,X,y)
```

```python
svc_model=SVC(kernel='poly',degree=2)  
svc_model.fit(X,y)
plot_svm_boundary(svc_model,X,y)
```

```python
svc_model=SVC(kernel='poly',degree=8)   #degree doesnt really matter for this dataset
svc_model.fit(X,y)
plot_svm_boundary(svc_model,X,y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'kernel':['linear','rbf'],'C':[0.01,0.1,1]}
```
