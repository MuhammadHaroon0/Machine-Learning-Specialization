```python
X=df.drop('class',axis=1)
```

```python
X
```

```python
X=pd.get_dummies(X)
```

```python
X
```

```python
y=df['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
abc=AdaBoostClassifier(n_estimators=1)
```

```python
abc.fit(X_train,y_train)
```

```python
preds=abc.predict(X_test)
```

```python
abc.feature_importances_
```

```python
abc.feature_importances_.argmax()
```

```python
X.columns[27]
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(classification_report(preds,y_test))
```

```python
error_rates=[]
for i in range(1,117):
    abc=AdaBoostClassifier(n_estimators=i)
    abc.fit(X_train,y_train)
    pred=abc.predict(X_test)
    error_rates.append(1-accuracy_score(y_test,pred))
```

```python
plt.plot(range(1,117),error_rates)
```
