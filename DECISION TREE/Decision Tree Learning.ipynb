```python
X=pd.get_dummies(data=df.drop('species',axis=1),drop_first=True)
```

```python
y=df['species']
```

```python
sns.heatmap(data=X) #not a good one for this data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#no need to scale the data in decision tree algo as it doesnt works on multiple features at same time
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
base_tree=DecisionTreeClassifier()
```

```python
base_tree.fit(X_train,y_train)
```

```python
preds=base_tree.predict(X_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,classification_report,confusion_matrix
```

```python
print(classification_report(preds,y_test))
```

```python
cm=confusion_matrix(y_test,preds)
```

```python
ConfusionMatrixDisplay(cm).plot()
```

```python
from sklearn.tree import plot_tree
```

```python
list(X.columns)
```

```python
plt.figure(figsize=(12,8),dpi=200)
plot_tree(base_tree,filled=True,feature_names=list(X.columns));
```

```python
base_tree.feature_importances_
```

```python
X_train.columns
```

```python
pd.DataFrame(index=X.columns,columns=['Feature Importance'],data=base_tree.feature_importances_).sort_values('Feature Importance')
```

```python
pruned_tree=DecisionTreeClassifier(max_depth=2)
```

```python
pruned_tree.fit(X_train,y_train)
```

```python
preds=pruned_tree.predict(X_test)
```

```python
print(classification_report(y_test,preds))
```

```python
cm=confusion_matrix(y_test,preds)
ConfusionMatrixDisplay(cm).plot()
```

```python
plt.figure(figsize=(12,8),dpi=200)
plot_tree(pruned_tree,filled=True,feature_names=list(X.columns));
```

```python
max_leaf_tree=DecisionTreeClassifier(max_leaf_nodes=3)
```

```python
max_leaf_tree.fit(X_train,y_train)
```

```python
preds=max_leaf_tree.predict(X_test)
```

```python
print(classification_report(y_test,preds))
```

```python
cm=confusion_matrix(y_test,preds)
ConfusionMatrixDisplay(cm).plot()
```

```python
plt.figure(figsize=(12,8),dpi=200)
plot_tree(max_leaf_tree,filled=True,feature_names=list(X.columns));
```

```python
tree=DecisionTreeClassifier(min_samples_leaf=100)
tree.fit(X_train,y_train)
preds=tree.predict(X_test)
plt.figure(figsize=(12,8),dpi=200)


plot_tree(tree,filled=True,feature_names=list(X.columns));
```

```python
cm=confusion_matrix(y_test,preds)
ConfusionMatrixDisplay(cm).plot()
```

```python
tree=DecisionTreeClassifier(criterion='entropy')
tree.fit(X_train,y_train)
preds=tree.predict(X_test)
plt.figure(figsize=(12,8),dpi=200)


plot_tree(tree,filled=True,feature_names=list(X.columns));
```

```python
cm=confusion_matrix(y_test,preds)
ConfusionMatrixDisplay(cm).plot()
```
